```python
adAttributed = pd.read_csv('c:\\kaggle\\adtracking\\attributed.csv', header = 0)

adAttributed.info()
adAttributed.head(10)
```

```python
print(pd.unique(adAttributed['app']))
#print(adAttributed['app'].value_counts())

v_count = adAttributed['app'].value_counts()
idStr = v_count.iloc[::].index
idVal = v_count.values

strIdx = '['
for ele in idStr:
    if len(strIdx) > 2:
        strIdx += ','    
    strIdx += str(int(ele))

strIdx += ']'
print(strIdx)

strVal = '['
for ele in idVal:
    if len(strVal) > 2:
        strVal += ','          
    strVal += str(int(ele))
strVal +=']'
        
print(strVal)
```
