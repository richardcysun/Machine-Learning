```python
import numpy as np

'''
Neighborhood           
MeadowV           88000
IDOTRR           103000
BrDale           106000
OldTown          119000
Edwards          121750
BrkSide          124300
Sawyer           135000
Blueste          137500
SWISU            139500
NAmes            140000
NPkVill          146000
Mitchel          153500
SawyerW          179900
Gilbert          181000
NWAmes           182900
Blmngtn          191000
CollgCr          197200
ClearCr          200250
Crawfor          200624
Veenker          218000
Somerst          225500
Timber           228475
StoneBr          278000
NoRidge          301500
NridgHt          315000
'''
def getNeighborMedian(nbr):
    if nbr == 'MeadowV':
        return 88000 
    elif nbr == 'IDOTRR':
        return 103000
    elif nbr == 'BrDale':
        return 106000  
    elif nbr == 'OldTown':
        return 119000   
    elif nbr == 'Edwards':
        return 121750
    elif nbr == 'BrkSide':
        return 124300
    elif nbr == 'Sawyer':
        return 135000
    elif nbr == 'Blueste':
        return 137500
    elif nbr == 'SWISU':
        return 139500
    elif nbr == 'NAmes':
        return 140000
    elif nbr == 'NPkVill':
        return 146000
    elif nbr == 'Mitchel':
        return 153500
    elif nbr == 'SawyerW':
        return 179900    
    elif nbr == 'Gilbert':
        return 181000   
    elif nbr == 'NWAmes':
        return 182900    
    elif nbr == 'Blmngtn':
        return 191000   
    elif nbr == 'CollgCr':
        return 197200
    elif nbr == 'ClearCr':
        return 200250
    elif nbr == 'Crawfor':
        return 200624
    elif nbr == 'Veenker':
        return 218000
    elif nbr == 'Somerst':
        return 225500
    elif nbr == 'Timber':
        return 228475
    elif nbr == 'StoneBr':
        return 278000  
    elif nbr == 'NoRidge':
        return 301500   
    elif nbr == 'NridgHt':
        return 315000  

def rmsle(y_true,y_pred):
   assert len(y_true) == len(y_pred)
   return np.square(np.log(y_pred + 1) - np.log(y_true + 1)).mean() ** 0.5
```

```python
from sklearn import metrics
from sklearn.cross_validation import KFold

X_fold = hsLnData[['OverallQual', 'Total_FlrSF', 'GrLivArea', 'Neighborhood_Median', 'Exter_QQ', 'GarageCars', 'GarageArea', \
            'TotalBsmtSF', 'Total_Bath', 'House_Freshment', 'Bsmt_QQ',\
            'KitchenQual','TotRmsAbvGrd']]
y_fold = hsLnData['SalePrice']

def run_KFold(regressor):
    kf = KFold(1366, n_folds=10, shuffle=True)
    outcomes = []
    
    fold = 0
    for train_index, test_index in kf:
        Xtrain, Xtest = X_fold.values[train_index], X_fold.values[test_index]
        ytrain, ytest = y_fold.values[train_index], y_fold.values[test_index]
        
        regressor.fit(Xtrain, ytrain)
        predictions = regressor.predict(Xtest)

        ytest = np.exp(ytest)
        predictions = np.exp(predictions)
            
        accuracy = rmsle(ytest, predictions)
        
        msg = 'Run ' + str(fold) + ': ' + str(accuracy)
        print msg
        outcomes.append(accuracy)
        fold += 1
        
    print("Mean: %s\n" % np.mean(outcomes))
```

```python
from xgboost import XGBRegressor
xg = XGBRegressor(max_depth=5, n_estimators=150)

run_KFold(xg)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(n_estimators = 150, max_depth=5)

run_KFold(gb)
```

```python
from sklearn.svm import SVR
svr = SVR(gamma=0.001, C=150, kernel='rbf')

run_KFold(svr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 150, max_depth=5)

run_KFold(rf)
```
