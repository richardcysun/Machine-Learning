```python
def getMsZoneCategory(zone):
    if zone == 'RL':
        return 5
    elif zone == 'RM':
        return 4
    elif zone == 'RH':
        return 3    
    elif zone == 'C (all)':
        return 2
    elif zone == 'FV':
        return 1

def getMsZoneIdx(zone):
    if zone == 'RL':
        return 5000
    elif zone == 'RM':
        return 4000
    elif zone == 'RH':
        return 3000 
    elif zone == 'C (all)':
        return 2000
    elif zone == 'FV':
        return 1000
    
def getQualityCategory(gQ):
    if gQ == 'NA':
        return 0
    elif gQ == 'Po':
        return 1
    elif gQ == 'Fa':
        return 2
    elif gQ == 'TA':
        return 3   
    elif gQ == 'Gd':
        return 4
    elif gQ == 'Ex':
        return 5

def getFoundationCategory(fnd):
    if fnd == 'PConc':
        return 5
    elif fnd == 'BrkTil':
        return 4
    elif fnd == 'CBlock':
        return 3
    elif fnd == 'Stone':
        return 2    
    elif fnd == 'Slab':
        return 1       
    elif fnd == 'Wood':
        return 0  
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
#panda is a read/write tool

train_file = "c:\\house\\train.csv"
test_file = "c:\\house\\test.csv"

# comma delimited is the default
hsData = pd.read_csv(train_file, header = 0)
hsData['MSZoning_ID'] = hsData['MSZoning'].apply(getMsZoneCategory)
hsData['GarageQual'] = hsData['GarageQual'].fillna('NA')
hsData['GarageQual_ID'] = hsData['GarageQual'].apply(getQualityCategory)
hsData['GarageYrBlt'] = hsData['GarageYrBlt'].fillna(hsData['YearBuilt'])
hsData['MasVnrArea'] = hsData['MasVnrArea'].fillna(0)
hsData['LotFrontage'] = hsData['LotFrontage'].fillna(hsData['LotFrontage'].median())
hsData['FireplaceQu'] = hsData['FireplaceQu'].fillna('NA')
hsData['FireplaceQu_ID'] = hsData['FireplaceQu'].apply(getQualityCategory)
hsData['FlrSF'] = hsData['1stFlrSF'] + hsData['2ndFlrSF']
hsData['PorchSF'] = hsData['OpenPorchSF'] + hsData['EnclosedPorch'] + hsData['3SsnPorch'] + hsData['ScreenPorch']
hsData['MSIdx'] = hsData['MSZoning'].apply(getMsZoneIdx) + hsData['MSSubClass']
hsData['Foundation_ID'] = hsData['Foundation'].apply(getFoundationCategory)

#print hsData['MSIdx'].loc[:20]
#hsData.drop('MSZoning', axis = 1)
#hsData.drop('GarageQual', axis = 1)
hsData.info()
hsData.describe()

hsTest = pd.read_csv(test_file, header = 0)

hsTest['MSZoning_ID'] = hsTest['MSZoning'].apply(getMsZoneCategory)
hsTest['GarageQual'] = hsTest['GarageQual'].fillna('NA')
hsTest['GarageCars'] = hsTest['GarageCars'].fillna(0)
hsTest['GarageArea'] = hsTest['GarageArea'].fillna(0)
hsTest['GarageQual_ID'] = hsTest['GarageQual'].apply(getQualityCategory)
hsTest['GarageYrBlt'] = hsTest['GarageYrBlt'].fillna(hsData['YearBuilt'])
hsTest['TotalBsmtSF'] = hsTest['TotalBsmtSF'].fillna(0)#need refine
hsTest['MasVnrArea'] = hsTest['MasVnrArea'].fillna(0)
hsTest['BsmtFinSF1'] = hsTest['BsmtFinSF1'].fillna(0)
hsTest['LotFrontage'] = hsTest['LotFrontage'].fillna(hsTest['LotFrontage'].median())
hsTest['FireplaceQu'] = hsTest['FireplaceQu'].fillna('NA')
hsTest['FireplaceQu_ID'] = hsTest['FireplaceQu'].apply(getQualityCategory)
hsTest['FlrSF'] = hsTest['1stFlrSF'] + hsTest['2ndFlrSF']
hsTest['PorchSF'] = hsTest['OpenPorchSF'] + hsTest['EnclosedPorch'] + hsTest['3SsnPorch'] + hsTest['ScreenPorch']
hsTest['MSZoning'] = hsTest['MSZoning'].fillna('RL')
hsTest['MSIdx'] = hsTest['MSZoning'].apply(getMsZoneIdx) + hsTest['MSSubClass']
hsTest['Foundation_ID'] = hsTest['Foundation'].apply(getFoundationCategory)

hsTest.info()
hsTest.describe()
```

```python
# The best so far
# Random Forest
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.externals import joblib

# No Ticket Info is better, it doesn't have useful information and it becomes noise
X = hsData[['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', 'FullBath', 'TotRmsAbvGrd', \
            'YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'MasVnrArea', 'Fireplaces', 'LotFrontage', \
            'FireplaceQu_ID', 'GarageQual_ID', 'FlrSF', 'OpenPorchSF', 'MSIdx', 'Foundation_ID']]
y = hsData['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y)

rf = RandomForestClassifier(n_estimators = 150, max_depth=5, oob_score=True)
rf.fit(X_train, y_train)

X_test_predict = rf.predict(X_test)

print(rf)
print("Number of True Prediction: %s\n" % metrics.accuracy_score(y_test, X_test_predict, normalize=False))
print("Score: %s\n" % rf.score(X_train, y_train))
#print(metrics.classification_report(y_test, X_test_predict))

Hs_Test = hsTest[['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', 'FullBath', 'TotRmsAbvGrd', \
            'YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'MasVnrArea', 'Fireplaces', 'LotFrontage', \
            'FireplaceQu_ID', 'GarageQual_ID', 'FlrSF', 'OpenPorchSF', 'MSIdx', 'Foundation_ID']]

Hs_test_predict = rf.predict(Hs_Test)

submit_file = 'c:\\house\\submission_rf.csv'
print submit_file
submission = pd.DataFrame({
        "Id": hsTest["Id"],
        "SalePrice": Hs_test_predict
    })

submission.to_csv(submit_file, index=False)
```
