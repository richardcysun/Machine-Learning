```python
from sklearn import metrics
from sklearn.cross_validation import KFold

X_fold = hsData[['FlrSF', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'Total_Bath', \
            'Qual_QC', 'HeatingQC', 'KitchenQual', 'Exter_OQ']]
y_fold = hsData['SalePrice']

def run_KFold(classifier):
    kf = KFold(1460, n_folds=10, shuffle=True)
    outcomes = []
    
    fold = 0
    for train_index, test_index in kf:
        Xtrain, Xtest = X_fold.values[train_index], X_fold.values[test_index]
        ytrain, ytest = y_fold.values[train_index], y_fold.values[test_index]
        
        classifier.fit(Xtrain, ytrain)
        predictions = classifier.predict(Xtest)
        accuracy = rmsle(ytest, predictions)
        
        msg = 'Run ' + str(fold) + ': ' + str(accuracy)
        print msg
        outcomes.append(accuracy)
        fold += 1
        
    print("Mean: %s\n" % np.mean(outcomes))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 150, max_depth=5)

run_KFold(rf)
```

```python
from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression()

run_KFold(LogReg)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators = 150, max_depth=5)

run_KFold(gb)
```
