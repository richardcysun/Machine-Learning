```python
import numpy as np

'''
tMeadowV median:88000.0, mean:98576.4705882, std:23491.0496097
tIDOTRR median:103000.0, mean:100123.783784, std:33376.7101169
tBrDale median:106000.0, mean:104493.75, std:14330.1764934
tOldTown median:119000.0, mean:128225.300885, std:52650.5831849
tEdwards median:121750.0, mean:128219.7, std:43208.6164587
tBrkSide median:124300.0, mean:124834.051724, std:40348.6892704
tSawyer median:135000.0, mean:136793.135135, std:22345.1291565
tBlueste median:137500.0, mean:137500.0, std:19091.883092
tSWISU median:139500.0, mean:142591.36, std:32622.9176792
tNAmes median:140000.0, mean:145847.08, std:33075.3454503
tNPkVill median:146000.0, mean:142694.444444, std:9377.31452911
tMitchel median:153500.0, mean:156270.122449, std:36486.6253336
tSawyerW median:179900.0, mean:186555.79661, std:55651.9978195
tGilbert median:181000.0, mean:192854.506329, std:35986.7790851
tNWAmes median:182900.0, mean:189050.068493, std:37172.2181065
tBlmngtn median:191000.0, mean:194870.882353, std:30393.229219
tCollgCr median:197200.0, mean:197965.773333, std:51403.6664382
tClearCr median:200250.0, mean:212565.428571, std:50231.5389928
tCrawfor median:200624.0, mean:210624.72549, std:68866.3954721
tVeenker median:218000.0, mean:238772.727273, std:72369.3179588
tSomerst median:225500.0, mean:225379.837209, std:56177.5558882
tTimber median:228475.0, mean:242247.447368, std:64845.6515493
tStoneBr median:278000.0, mean:310499.0, std:112969.67664
tNoRidge median:301500.0, mean:335295.317073, std:121412.65864
tNridgHt median:315000.0, mean:316270.623377, std:96392.5449544
'''
def getSortedNeighborCategory(nbr):
    if nbr == '':
        return 0
    elif nbr == 'MeadowV':
        return 1    
    elif nbr == 'IDOTRR':
        return 2    
    elif nbr == 'BrDale':
        return 3    
    elif nbr == 'OldTown':
        return 4   
    elif nbr == 'Edwards':
        return 5    
    elif nbr == 'BrkSide':
        return 6 
    elif nbr == 'Sawyer':
        return 7 
    elif nbr == 'Blueste':
        return 8     
    elif nbr == 'SWISU':
        return 9
    elif nbr == 'NAmes':
        return 10  
    elif nbr == 'NPkVill':
        return 11   
    elif nbr == 'Mitchel':
        return 12 
    elif nbr == 'SawyerW':
        return 13    
    elif nbr == 'Gilbert':
        return 14   
    elif nbr == 'NWAmes':
        return 15    
    elif nbr == 'Blmngtn':
        return 16    
    elif nbr == 'CollgCr':
        return 17  
    elif nbr == 'ClearCr':
        return 18    
    elif nbr == 'Crawfor':
        return 19    
    elif nbr == 'Veenker':
        return 20
    elif nbr == 'Somerst':
        return 21    
    elif nbr == 'Timber':
        return 22  
    elif nbr == 'StoneBr':
        return 23    
    elif nbr == 'NoRidge':
        return 24    
    elif nbr == 'NridgHt':
        return 25    

def rmsle(y_true,y_pred):
   assert len(y_true) == len(y_pred)
   return np.square(np.log(y_pred + 1) - np.log(y_true + 1)).mean() ** 0.5
```

```python
from sklearn import metrics
from sklearn.cross_validation import KFold

X_fold = hsData[['FlrSF', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'Total_Bath', \
            'Qual_QC', 'HeatingQC', 'KitchenQual', 'Exter_OQ']]
y_fold = hsData['SalePrice']

def run_KFold(classifier):
    kf = KFold(1460, n_folds=10, shuffle=True)
    outcomes = []
    
    fold = 0
    for train_index, test_index in kf:
        Xtrain, Xtest = X_fold.values[train_index], X_fold.values[test_index]
        ytrain, ytest = y_fold.values[train_index], y_fold.values[test_index]
        
        classifier.fit(Xtrain, ytrain)
        predictions = classifier.predict(Xtest)
        accuracy = rmsle(ytest, predictions)
        
        msg = 'Run ' + str(fold) + ': ' + str(accuracy)
        print msg
        outcomes.append(accuracy)
        fold += 1
        
    print("Mean: %s\n" % np.mean(outcomes))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 150, max_depth=5)

run_KFold(rf)
```

```python
from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression()

run_KFold(LogReg)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators = 150, max_depth=5)

run_KFold(gb)
```
