```python
def fillAgeBySocRank(soc_rank):
    if soc_rank == 0:
        return 31
    elif soc_rank == 1:
        return 22
    elif soc_rank == 2:
        return 3.5
    elif soc_rank == 3:
        return 42
    elif soc_rank == 4:
        return 36
    elif soc_rank == 5:
        return 35.5
    else:
        return 31
    
def getEmbSexCategory(embsex):
    if embsex == 'Smale':
        return 1
    elif embsex == 'Sfemale':
        return 2
    elif embsex == 'Cmale':
        return 3
    elif embsex == 'Cfemale':
        return 4
    elif embsex == 'Qmale':
        return 5
    elif embsex == 'Qfemale':
        return 6    

def getCabinCategory(cabin):
    for ele in cabin:
        if ele == 'A':
            return 1
        elif ele == 'B':
            return 2
        elif ele == 'C':
            return 3
        elif ele == 'D':
            return 4
        elif ele == 'E':
            return 5
        elif ele == 'F':
            return 6
        elif ele == 'G':
            return 7
        elif ele == 'T':
            return 8
        else:
            return 0
        
def getPclassCategory(pcl):
    if pcl == 1:
        return 'PCL1'
    elif pcl == 2:
        return 'PCL2'
    else:
        return 'PCL3'
    
def getPclSexCategory(pclsex):
    if pclsex == 'PCL1male':
        return 0
    elif pclsex == 'PCL1female':
        return 1
    elif pclsex == 'PCL2male':
        return 2
    elif pclsex == 'PCL2female':
        return 3
    elif pclsex == 'PCL3male':
        return 4
    elif pclsex == 'PCL3female':
        return 5    
```

```python
# The best so far
# Random Forest
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.externals import joblib

# No Ticket Info is better, it doesn't have useful information and it becomes noise
X = tiData[['Pclass', 'Age', 'Female', 'Fam_Grp', 'Fare', 'Embarked', 'Emb_Sex', 'Soc_Rank', 'Cabin_ID']]
y = tiData['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y)

rf = RandomForestClassifier(n_estimators = 170, max_depth=5, oob_score=True, n_jobs=-1)
rf.fit(X_train, y_train)

X_test_predict = rf.predict(X_test)

print(rf)
print("Number of True Prediction: %s\n" % metrics.accuracy_score(y_test, X_test_predict, normalize=False))
print("Score: %s\n" % rf.score(X_test, y_test))
print(metrics.classification_report(y_test, X_test_predict))

Ti_test = tiTest[['Pclass', 'Age', 'Female', 'Fam_Grp', 'Fare', 'Embarked', 'Emb_Sex', 'Soc_Rank', 'Cabin_ID']]

Ti_test_predict = rf.predict(Ti_test)
#print(Ti_test_predict)

# compare with the almost data
almost_perfect_file = "c:\\kaggle\\titanic\\almost_perfect.csv"
almostPref = pd.read_csv(almost_perfect_file, header = 0)
alpSurv = almostPref['Survived']

idx = 0
okay = 0
for ti_pred in Ti_test_predict:
    if ti_pred == alpSurv.iloc[idx]:
        okay += 1
    idx += 1
        
print(okay)
print(idx)

submit_file = 'c:\\kaggle\\titanic\\submission_rf_' + str(okay) + '.csv'
print(submit_file)
submission = pd.DataFrame({
        "PassengerId": tiTest["PassengerId"],
        "Survived": Ti_test_predict
    })

submission.to_csv(submit_file, index=False)

if okay > best_okay:
    best_okay = okay
    model_file = 'c:\\kaggle\\titanic\\rf_' + str(okay) + '.pkl'
    print(model_file)
    joblib.dump(rf, model_file) 
```
